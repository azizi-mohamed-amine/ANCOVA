```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pingouin import ancova
```

```python
df = pd.read_excel(r'C:\Users\amin\Downloads\SNAP_ADHD_all.xlsx')
df
```

```python
df1 = df[['Subject', 'ADHDInattention', 'ADHDHyperImpulsive', 'ADHDCombined', 'ConnersIndex', 'InattentionOveractivity', 'InattentionDomain']]
df1
```

```python
v=len(df1.columns)
dg = df1.sum(axis=1)/v
dg
```

```python
dd = {'average':  dg,
        'Group (1=ADHD, 2=HC)': df['Group (1=ADHD, 2=HC)'],
     'HyperactivityImpulsivityDomain':df['HyperactivityImpulsivityDomain']}

dd = pd.DataFrame (dd, columns = ['average','Group (1=ADHD, 2=HC)','HyperactivityImpulsivityDomain'])
dd
```

```python
ancova(data=dd, dv='average', covar='HyperactivityImpulsivityDomain', between='Group (1=ADHD, 2=HC)')
```
